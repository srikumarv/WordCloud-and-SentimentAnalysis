```python
#!/usr/bin/python
from nltk.book import *
from wordcloud import WordCloud
from nltk.corpus import stopwords
from wordcloud import STOPWORDS
import MySQLdb as mdb
import pandas

#-----------Functions Defined -----------------------------------#


#-------------Getting DB Connection Function - Returns Connection --------#
def get_db_connection(config):
    #Creating connection to MYSQL DB
    try:
        con = mdb.connect(host = config["host"], 
                      user = config["user"], 
                      passwd = config["passwd"], 
                      db = config["db"], 
                      charset='utf8', use_unicode=True);
    except :
            print("Oops!  DB Connection Error!!!")

    return con

#--------------Get Cursor for seleted DB ---------------------#

def get_cursor(con):
    try:
    #DB  Name defined

        cur = con.cursor(mdb.cursors.DictCursor)
    except :
            print("Oops!  Cursor Fetch Error!!!")
            
    return cur

def close_db_connection(con):
    try:
    #DB  Name defined

        con.close()
    except :
            print("Oops!  Cannot close connection!!!")

#-------------End of Functions ---------------------------------#     


config = {}
execfile("config_1.txt", config)

viztext= "bad worse worst good better"

sw = stopwords.words('english')
sw.extend(STOPWORDS)


#----------------GETDATA FROM DB-------------------------
try:
#DB  Name defined

    con = get_db_connection(config)
    cur = get_cursor(con)
    #cur = con.cursor(mdb.cursors.DictCursor)
except :
        print("Oops!  Cursor Fetch Error!!!")

cur.execute(config['poswordsforhillary'])

poswords  = ""
for row in cur:
    
    poswords +=row["word"]
    poswords += " " 
        
cur.close()
close_db_connection(con)



#------------end get data----------------------------------

#-----------wordcloud for every candidate displayed -----------------

wordcloud = WordCloud(stopwords=sw, background_color="red").generate(poswords)

%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
